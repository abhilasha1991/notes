# join

```python
left =pd.DataFrame({'A':['A0','A1','A2'],
                   'B':['B0','B1','B2']}, index=['K0','K1','K2'])

right =pd.DataFrame({'C':['C0','C2','C3'],
                    'D':['D0','D2','D3']},
                   index=['K0','K2','K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)#left will not change right should change according to left
```

```python
left.join(right, how='outer')#so for full join
```

```python
right.join(left)
```

```python
right.join(left,how='outer')
```

```python
#defines a function
def salary_hike(x):
    if (x > 3000):
        return np.ceil(x + (x *.30))
    else:
        return np.ceil(x * 1.5)
```

```python
df['new_salary'] = df['salary'].apply(salary_hike)
print(df)
```

```python
df['name_length']= df['emp'].apply(len)
print(df)
```
